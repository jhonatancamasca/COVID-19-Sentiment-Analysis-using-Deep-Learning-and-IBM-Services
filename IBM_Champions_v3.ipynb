#Limpieza de la data

```python
def remove_urls(row): 
    text=row['text']
    url_pattern = re.compile(r'https?://\S+|www\.\S+')
    a=url_pattern.sub(r'', text)
    return a
```

```python
def demojir(row):
    text=row['text']
    text=demoji.replace(text)
    return text
```

```python
def remove_user(row):
    text=row['text']
    r = re.findall("@[\w]*", text)
    for i in r:
        text = re.sub(i, '', text)
    return text
```

```python
def remove_hastag(row):
    text=row['text']
    r = re.findall("#[\w]*", text)
    for i in r:
        text = re.sub(i, '', text)
    return text
    
```

```python
def remove_rt(row):
    text=row['text']
    r = re.findall('RT :', text)
    for i in r:
        text = re.sub(i, '', text)
    return text.lower()
```

```python
def signos(row):

    sentences=row['text']
    text_string=re.findall('[\w]+',sentences)
    cad=' '.join(text_string)
    cad=unidecode.unidecode(cad)
    cad2=cad.translate(str.maketrans("",'',a))
    cad3=cad2.split()
    cad4=' '.join(cad3)
    
    return cad4
```

```python
data['text'] = data.apply(demojir, axis=1)
data['text'] = data.apply(remove_user, axis=1)
data['text'] = data.apply(remove_hastag, axis=1)
data['text'] = data.apply(remove_rt, axis=1)
data['text'] = data.apply(remove_urls, axis=1)
data['text'] = data.apply(signos, axis=1)
```

```python
nv_format=[]
data1=list(data['text'])
for i in data1:
    nv_format.append([i])

#Labeling
labels = ['negative','positive']
valores=[]
for i in range(len(nv_format)):
    
    seq = tokenizer.texts_to_sequences(nv_format[i])
    #print(i)
    padded = sequence.pad_sequences(seq, maxlen=max_len)
    #print (padded)
    pred = model1.predict(padded)
    pred1 = model2.predict(padded) 
    pred2 = model3.predict(padded) 
    l1=labels[np.argmax(pred)]
    l2=labels[np.argmax(pred1)]
    l3=labels[np.argmax(pred2)]
    arr=[l1,l2,l3]
    if arr.count('negative')>arr.count('positive'):
        a='negative'
    else:
        a='positive'
    
    valores.append([a,nv_format[i][0]])
```
